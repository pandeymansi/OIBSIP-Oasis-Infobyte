# Importing Libraries

```python
# rename columns
data.columns = ['States','Date','Frequency','Estimated Unemployment Rate',
                'Estimated Employed','Estimated Labour Participation Rate',
                'Region']
```

```python
# analysing top rows of dataset
data.head()
```

# Checking the correlation between the feature of dataset

```python
# Set the plot style and figure size
plt.style.use('seaborn-dark')
plt.figure(figsize=(10, 8))

# Select only numerical columns for correlation calculation
numerical_data = data.select_dtypes(include=['number'])

# Compute the correlation matrix
correlation_matrix = numerical_data.corr()

# Define a custom colormap (Blues) and plot the heatmap
ax = sns.heatmap(correlation_matrix, annot=True, linewidth=2, cmap='Blues',
                 cbar_kws={'orientation': 'vertical', 'shrink': 0.75})

# Customize tick parameters and labels
ax.tick_params(size=8, color='black', labelsize=12, labelcolor='darkblue')

# Add a title and axis labels for clarity
plt.title('Correlation Heatmap of Numerical Features', fontsize=18, color='navy')
plt.xlabel('Feature', fontsize=14, color='black')
plt.ylabel('Feature', fontsize=14, color='black')

# Rotate tick labels for better readability
plt.xticks(rotation=45)
plt.yticks(rotation=0)

# Show the heatmap
plt.show()
```

# Estimated number of employee according to different region of india

```python
# plotting histplot

data.columns=['States','Date','Frequency','Estimated Unemployment Rate',
                'Estimated Employed','Estimated Labour Participation Rate',
                'Region']
plt.title('Indian Unemployment')
sns.histplot(x='Estimated Employed',hue='Region',data=data)
plt.show()
```

# Unemployment rate according to different regions of india

```python
# plotting histplot

plt.figure(figsize=(10,8))
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Unemployment Rate",hue='Region',data=data)
plt.show()
```

# Dashboard to analyze the unemployment rate of each Indian state

```python
# dashboard to analyze the unemployment rate of each indian state

# plotting sunburst

unemployment = data[['States','Region','Estimated Unemployment Rate']]

# Group by Region and State to aggregate unemployment rates for non-leaf nodes
unemployment_agg = unemployment.groupby(['Region', 'States'])['Estimated Unemployment Rate'].sum().reset_index()

figure = px.sunburst(unemployment_agg, path=['Region','States'],
                     values='Estimated Unemployment Rate',
                     width=700,height=600, color_continuous_scale='RdY1Gn',
                     title="Unemployment Rate in India")
figure.show()
```
